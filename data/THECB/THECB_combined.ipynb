```python
import pandas as pd

df2019 = pd.read_excel("HSGradsEnrolledDistrict2019.xlsx", header=8)
df2020 = pd.read_excel("HSGradsEnrolledDistrict2020.xlsx", header=8)
df2021 = pd.read_excel("HSGradsEnrolledDistrict2021.xlsx", header=8)
df2022 = pd.read_excel("HSGradsEnrolledDistrict2022.xlsx", header=8)
df2023 = pd.read_excel("HSGradsEnrolledDistrict2023.xlsx", header=8)
df2024 = pd.read_excel("HSGradsEnrolledDistrict2024.xlsx", header=8)
```

```python

dfs = {
    2019: df2019,
    2020: df2020,
    2021: df2021,
    2022: df2022,
    2023: df2023,
    2024: df2024,
}

combined = pd.concat(
    [
        df.assign(year=year)
        for year, df in dfs.items()
    ],
    ignore_index=True
)
```
